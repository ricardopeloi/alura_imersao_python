# Leitura de dados

```python
def convert_google_sheet_url(url):
    # Regular expression to match and capture the necessary part of the URL
    pattern = r'https://docs\.google\.com/spreadsheets/d/([a-zA-Z0-9-_]+)(/edit#gid=(\d+)|/edit.*)?'

    # Replace function to construct the new URL for CSV export
    # If gid is present in the URL, it includes it in the export URL, otherwise, it's omitted
    replacement = lambda m: f'https://docs.google.com/spreadsheets/d/{m.group(1)}/export?' + (f'gid={m.group(3)}&' if m.group(3) else '') + 'format=csv'

    # Replace using regex
    new_url = re.sub(pattern, replacement, url)

    return new_url

base_ticker = pd.read_csv(convert_google_sheet_url("https://docs.google.com/spreadsheets/d/1FO7R8HkhqqHfVgup5-52DTDKT_roiVpBARNTlgxchCo/edit#gid="+str(726331890)))
base_ticker

base_total_de_acoes = pd.read_csv(convert_google_sheet_url("https://docs.google.com/spreadsheets/d/1FO7R8HkhqqHfVgup5-52DTDKT_roiVpBARNTlgxchCo/edit#gid="+str(1775453746)))
base_total_de_acoes

base_segmento_idade = pd.read_csv(convert_google_sheet_url("https://docs.google.com/spreadsheets/d/1FO7R8HkhqqHfVgup5-52DTDKT_roiVpBARNTlgxchCo/edit#gid="+str(1098796981)))
base_segmento_idade

base_principal = pd.read_csv(convert_google_sheet_url("https://docs.google.com/spreadsheets/d/1FO7R8HkhqqHfVgup5-52DTDKT_roiVpBARNTlgxchCo/edit#gid="+str(0)))
base_principal
```

# Tratamento dos dados

## Valor inicial

```python
base_principal_tratada = base_principal[["Ativo", "Data", "Último (R$)", "Var. Dia (%)"]].copy()

base_principal_tratada.columns = ["Ativo", "Data", "Valor final", "Variação percentual"]


base_principal_tratada["Valor inicial"] = base_principal_tratada["Valor final"].str.replace(",", ".").astype(float)/(1+(base_principal_tratada["Variação percentual"].str.replace(",", ".").astype(float)/100))
base_principal_tratada["Valor final"] = base_principal_tratada["Valor final"].str.replace(",", ".").astype(float)
base_principal_tratada["Variação percentual"] = base_principal_tratada["Variação percentual"].str.replace(",", ".").astype(float)/100

base_principal_tratada
```

## Total de ações e Variação total

```python
display(base_total_de_acoes.tail())
display(base_principal["Qtd ações"].str.replace(".", "").astype(float).sum())
```

```python
base_principal_tratada_total_acoes = base_principal_tratada.set_index("Ativo").join(base_total_de_acoes.set_index("Código")).reset_index()

base_principal_tratada_total_acoes["Qtde. Teórica"] = base_principal_tratada_total_acoes["Qtde. Teórica"].str.replace(".", "").astype(float)

display(base_principal_tratada_total_acoes["Qtde. Teórica"].sum())

base_principal_tratada_total_acoes["Variação total R$"] = (base_principal_tratada_total_acoes["Valor final"]-base_principal_tratada_total_acoes["Valor inicial"])*base_principal_tratada_total_acoes["Qtde. Teórica"]
display(base_principal_tratada_total_acoes)
```

## Resultado (com e sem ChatGPT)

```python
base_principal_tratada_total_acoes_resultado = base_principal_tratada_total_acoes.copy()

base_principal_tratada_total_acoes_resultado.loc[base_principal_tratada_total_acoes_resultado["Variação total R$"] > 0, "Resultado"] = "Subiu"
base_principal_tratada_total_acoes_resultado.loc[base_principal_tratada_total_acoes_resultado["Variação total R$"] < 0, "Resultado"] = "Desceu"
base_principal_tratada_total_acoes_resultado.loc[base_principal_tratada_total_acoes_resultado["Variação total R$"] == 0, "Resultado"] = "Estável"

base_principal_tratada_total_acoes_resultado
```

```python
# prompt: Usando o DataFrame base_principal_tratada_total_acoes_resultado: Generate a line of code in python pandas that can fill a column of a dataframe based on rules of another column, such as a number bigger than zero (growth), less than zero (decrease) or equal to zero (stable)

base_principal_tratada_total_acoes_resultado['Resultado ChatGPT'] = np.where(base_principal_tratada_total_acoes_resultado['Variação percentual'] > 0, 'Subiu', np.where(base_principal_tratada_total_acoes_resultado['Variação percentual'] < 0, 'Desceu', 'Estável'))

base_principal_tratada_total_acoes_resultado
sum(base_principal_tratada_total_acoes_resultado["Resultado ChatGPT"] == base_principal_tratada_total_acoes_resultado["Resultado"])
```
